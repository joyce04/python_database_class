## value_counts()

```python
club_count = players.club.value_counts()
club_count.head()
```

```python
def get_counts(club_list):
    counts = {}
    for x in club_list:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
get_counts(players.club.tolist())
```
