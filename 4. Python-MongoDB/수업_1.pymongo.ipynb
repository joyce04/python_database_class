## randint

```python
people = ['grace', 'kyle', 'kim', 'Lee']
city = ['seoul', 'newyork', 'london']
department = ['it', 'finance', 'HR']
```

```python
from random import *
```

```python
for person in people:
    emp_col.insert({'name': person,
                  'age': randint(20, 60),
                  'address': city[randint(0, 2)],
                  'department': department[randint(0, 2)]})
```

```python
list(emp_col.find())
```

```python
football_db = client.Football
```

```python
football_db
```

```python
wc_c = football_db.wc_match
```

```python
wc_c
```

```python
cursor = wc_c.find()
```

```python
list(cursor)[0]
```

```python
# aggregate
```

```python
home_goals = list(wc_c.aggregate([
    {'$group':{
        '_id': '$home_team.country',
        'home_goal':{'$sum': '$home_team.score'}
    }},
    {'$sort': {'home_goal':-1}}
]))
```

```python
home_goals
```

```python
away_goals = list(wc_c.aggregate([
    {'$group':{
        '_id': '$away_team.country',
        'away_goal':{'$sum': '$away_team.score'}
    }},
    {'$sort': {'away_goal':-1}}
]))
```

```python
away_goals
```

```python
import pandas as pd
```

```python
home_df = pd.DataFrame(home_goals)
away_df = pd.DataFrame(away_goals)
```

```python
home_df.head()
```

```python
away_df.head()
```

```python
total = pd.merge(home_df, away_df)
```

```python
total.head()
```

```python
total['total_goals'] = total.home_goal + total.away_goal
```

```python
total.head()
```

```python
total.sort_values('total_goals', ascending=True)
```
